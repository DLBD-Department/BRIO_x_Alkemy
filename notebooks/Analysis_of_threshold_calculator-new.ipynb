```python
from src.bias.threshold_calculator import threshold_calculator as tc
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(2, 10)
y1 = tc(A1="high", A2=x, A3=30)
y2 = tc(A1="high", A2=x, A3=100)
y3 = tc(A1="high", A2=x, A3=500)
fig, ax = plt.subplots()
ax.plot(x, y1, label="A3=30")
ax.plot(x, y2, label="A3=100")
ax.plot(x, y3, label="A3=500")
ax.set_title('Threshold with different A2 values (A1="high")', size=11)
ax.set_xlabel("A2")
ax.set_ylabel("epsilon")
plt.legend()
plt.show()
```

```python
x = np.arange(2, 10)
y1 = tc(A1="low", A2=x, A3=30)
y2 = tc(A1="low", A2=x, A3=100)
y3 = tc(A1="low", A2=x, A3=500)
fig, ax = plt.subplots()
ax.plot(x, y1, label="A3=30")
ax.plot(x, y2, label="A3=100")
ax.plot(x, y3, label="A3=500")
ax.set_title('Threshold with different A2 values (A1="low")', size=11)
ax.set_xlabel("A2")
ax.set_ylabel("epsilon")
plt.legend()
plt.show()
```

```python
x = np.arange(2, 10)
y = np.linspace(30, 500, 1000)
X, Y = np.meshgrid(x, y)
Z = tc(A1="high", A2=X, A3=Y)
fig, ax = plt.subplots()
plt.xlim([1, 10])
plt.ylim([30, 550])
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=True, fontsize=10)
ax.set_title('threshold calculator contour plot (A1 = high)')
ax.set_xlabel("A2")
ax.set_ylabel("A3")
```
